```python
import numpy as np
import pandas as pd
## Wczytanie pliku
cars = pd.read_csv('samochody1tys.csv')
```

```python
## WYŚWIETLANIE DANYCH
## Wyświetlenie zawartości pliku
cars
## Wyświetlenie kolumn
cars.columns
## Wypisanie pierwszych 10 wierszy
cars.head(10)
## Wyświetlenie statystyk danych
cars.describe()
## Wyświetlenie wartości danej kolumny
cars['marka']
## Wyświetlenie wartości z konkretnych kolumn i wierszy
cars[['model','przebieg']][0:5]
```

```python
## SORTOWANIE
## Sortowanie danych po kolumnie (rosnąco)
cars.sort_values('cena')
cars.sort_values(by='cena', ascending=True)
```

```python
## FILTROWANIE
## Wyświetlenie wartości z kilku wierszy
cars[0:2]
## Selekcja wierszy spełniających dany warunek
cars.loc[cars['rodzaj_silnika']=="Diesel"]
cars.loc[cars['wojewodztwo'].str.contains('Pomorskie')]
cars.loc[~cars['wojewodztwo'].str.contains('Pomorskie')]
cars.loc[(cars['marka']=='Nissan') | (cars['rodzaj_silnika']=='Diesel')]
## Wyświetlenie konkretnego rekordu
cars.iloc[0]
## Wyświetlenie wierszy z danego przedziału wierszy oraz kolumn
cars.iloc[3:5, 0:5]
## Wykorzystanie metody isin() do filtrowania danych
cars[cars['rok_produkcji'].isin(['2008','2010'])]
```

```python
## GRUPOWANIE 
## Pogrupowanie danych ze zliczeniem wyników
cars.groupby(['rodzaj_silnika']).count()
## Pogrupowanie danych z sumowaniem wyników
cars.groupby(['rodzaj_silnika']).sum()
```

```python
## DODATKOWE OPERACJE
## Usuwanie danych z podanej kolumny 
# (wynik operacji należy przypisac do wczytanego pliku)
cars.drop(['rok_produkcji'],axis=1).head(1000)
## Dodanie nowej kolumny
cars['stosunek_przebiegu_do_ceny']=cars.przebieg/cars.cena
cars
```
