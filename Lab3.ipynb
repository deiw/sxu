```python
import matplotlib.pyplot as plt
import numpy as np

# wykres funkcji liniowej, kwadratowej, trzeciego rzędu oraz funkcji sinusoidalnej
# Stworzenie zakresu dla danych wykresu
x = np.linspace(0, 1, 100)
# Wykres funkcji liniowej
plt.plot(x, x, label='liniowa')
# Wykres funkcji kwadratowej
plt.plot(x, x**2, label='kwadratowa')
# Wykres funkcji sześciennej
plt.plot(x, x**3, label='sześcienna')
# Wykres funkcji sinusoidalnej
plt.plot(x, np.sin(x), label='sinusoidalna')
# Dodanie tytułu wykresu
plt.title("Wykresy")
# Dodanie opisu osi x
plt.xlabel('X')
# Dodanie opisu osi y
plt.ylabel('Y')
# Pokazanie legendy
plt.legend()
# Wyświetlenie wykresu
plt.show()
```

```python
import pandas as pd
## Wczytanie pliku
cars = pd.read_csv('samochody1tys.csv')
```

```python
# Wykres słupkowy
# Średnia cena BMW
bmw = cars.loc[cars['marka']=="BMW", 'cena'].mean()
# Średnia cena AUDI
audi = cars.loc[cars['marka']=="Audi", 'cena'].mean()
# Dane do wykresu
dane = [bmw, audi]
# Pozycja wyświetlania słupków na wykresie
x_pos = np.arange(len(dane))
# Stworzenie wykresu słupkowego
ax = plt.subplot()
ax.bar(x_pos,dane, align='center')
# Opisy wyświetlanych danych
marki = ['BMW', 'Audi']
# Przypisanie pozycji opisów do słupków
ax.set_xticks(x_pos)
# Dodanie opisów słupków
ax.set_xticklabels(marki)
# Dodanie opisu osi y
plt.ylabel('Cena')
# Dodanie opisu osi x
plt.xlabel('Marka')
# Wyświetlenie wykresu
plt.show()
```

```python
# Wykres kołowy
# Dane do wykresu (liczba samochodów o określonej marce)
bmw = cars.loc[cars['marka']=="BMW", 'id'].count()
audi = cars.loc[cars['marka']=="Audi", 'id'].count()
ford = cars.loc[cars['marka']=="Ford", 'id'].count()
opel = cars.loc[cars['marka']=="Opel", 'id'].count()
# Utworzenie wykresu
fig, ax = plt.subplots();
# Przypisanie danych do zmiennej
dane = [bmw, audi, ford, opel]
# Tabelki z opisami do wykresu
opis = ['Bmw', 'Audi', 'Ford', 'Opel']
# Stworzenie wykresu kołowego
ax.pie(dane, labels=opis, autopct='%1.1f%%')
# Dodanie tytułu do wykresu
ax.set_title('Liczba samochodów')
# Wyświetlenie wykresu
plt.show()
```

```python
# Histogram
# Dane
dane = cars[cars.marka == 'Nissan'].groupby('wojewodztwo')['marka'].count()
# Stworzenie histogramu
fig, ax = plt.subplots()
ax.hist(dane, 10)
# Dodanie tytułu
ax.set_title('Liczba nissanów w poszczególnych województwach')
# Opis osi y
ax.set_ylabel('Liczba')
# Wyświetlenie wykresu
plt.show()
```
